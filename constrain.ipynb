```python
from datetime import datetime

def normalize_ops_event(raw):
    return {
        "timestamp": datetime.fromisoformat(raw["timestamp"]),
        "source": "ops",
        "entity": raw["host"],
        "anomaly_score": raw["anomaly_score"],
        "signal": "ops_anomaly"
    }

def normalize_security_event(raw):
    return {
        "timestamp": datetime.fromisoformat(raw["timestamp"]),
        "source": "security",
        "entity": raw["src_ip"],
        "anomaly_score": raw["anomaly_score"],
        "signal": "security_anomaly"
    }
```
