```python
features = [
    "Total Length of Fwd Packets",  # bytes per flow
    "Flow Packets/s",               # traffic rate
    "Flow Duration",                # session length
    "Destination Port",             # service / protocol behavior
    "Total Fwd Packets",            # connection intensity
    "SYN Flag Count"                # connection initiation behavior
]
```

```python
import numpy as np

df = df.replace([np.inf, -np.inf], np.nan)
df = df.dropna(subset=features)
```

```python
benign = df[df["Label"] == "BENIGN"]
X_train = benign[features]
```

```python
from sklearn.ensemble import IsolationForest

model = IsolationForest(
    n_estimators=200,
    contamination=0.05,
    random_state=42,
    n_jobs=-1
)

model.fit(X_train)
```

```python
raw_scores = -model.score_samples(df[features])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df["anomaly_score"] = scaler.fit_transform(
    raw_scores.reshape(-1, 1)
)
```

```python
def get_signal(score):
    if score > 0.7:
        return "unusual_traffic_pattern"
    elif score > 0.4:
        return "slightly_unusual"
    else:
        return "normal_traffic"

df["signal"] = df["anomaly_score"].apply(get_signal)
```

```python
output = {
    "source": "security",
    "anomaly_score": round(df.iloc[0]["anomaly_score"], 2),
    "signal": df.iloc[0]["signal"]
}

output
```

```python
import pandas as pd

# Generate a TRUE normal sample from training data
synthetic_normal = X_train.sample(1, random_state=42).copy()

# Add very small noise (optional)
synthetic_normal["Flow Packets/s"] *= 1.05
synthetic_normal["Total Fwd Packets"] *= 1.1

```

```python
synthetic_unusual = pd.DataFrame([{
    "Total Length of Fwd Packets": 80000,   # larger transfer
    "Flow Packets/s": 200,                  # bursty
    "Flow Duration": 30000,                 # short-lived
    "Destination Port": 8080,               # uncommon port
    "Total Fwd Packets": 600,
    "SYN Flag Count": 4
}])
```

```python
synthetic_attack = pd.DataFrame([{
    "Total Length of Fwd Packets": 5000000, # very large volume
    "Flow Packets/s": 5000,                 # flood-like
    "Flow Duration": 5000,                  # very short
    "Destination Port": 23,                 # telnet
    "Total Fwd Packets": 10000,
    "SYN Flag Count": 500                   # scanning behavior
}])
```

```python
train_raw_scores = -model.score_samples(X_train)
```

```python
p95 = np.percentile(train_raw_scores, 95)
p99 = np.percentile(train_raw_scores, 99)
```

```python
def score_synthetic_raw(sample_df):
    raw_score = -model.score_samples(sample_df[features])[0]

    if raw_score >= p99:
        signal = "unusual_traffic_pattern"
    elif raw_score >= p95:
        signal = "slightly_unusual"
    else:
        signal = "normal_traffic"

    return {
        "source": "security",
        "raw_anomaly_score": float(raw_score),
        "signal": signal
    }
```

```python
print(score_synthetic_raw(synthetic_normal))
print(score_synthetic_raw(synthetic_unusual))
print(score_synthetic_raw(synthetic_attack))
```
